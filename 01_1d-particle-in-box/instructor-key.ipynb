# Lesson: Particle in a 1-Dimensional Box
---
In this lesson, we will first solve the wave equation for a particle confined to an infinite 1-dimensional square well and then explore the properties of the particle wavefunctions. 

## Motivation

Imagine an infinitely narrow and perfectly conducting wire of finite length $L$ surrounded by a perfectly insulating vacuum, onto which an individual electron has been placed. Because the wire is surrounded by a perfect insulator, the electron cannot venture beyond the ends of the wire; however because the wire itself is perfectly conducting, the electron explore the full length of the wire with zero resistance to its flow. To explore the properties of this electron-on-a-wire, we can construct a quantum mechanical _model system_ which simplifies the picture while recovering all of the essential physics of this real system.

## I. Introduction & Background

Quantum mechanical systems are subject to the time-independent wave equation (TIWE)
$$\hat{H}\Psi = E\Psi,$$
where 
* $\hat{H}$ is the _Hamiltonian operator_ corresponding to the total energy of the system;
* $\Psi$ is the _wavefunction_, which contains all knowable information about the system; and
* $E$ is the total energy of the system described by the wavefunction $\Psi$.
Before we can solve for the wavefunction $\Psi$ describing our electron-on-a-wire, we must first define a model system representing its essential features and construct the Hamiltonian $\hat{H}$ which generates the TIWE for our model system.

### Defining the Model System

To explore the real electron-on-a-wire system, let us consider a model system where a single particle of mass $m$ is constrained to move in only one spatial dimension, which without loss of generality we will call $x$. The Hamiltonian operator $\hat{H}$ for this system corresponds to the total energy of our particle, which in general is given by 
$$\hat{H} = \hat{T} + \hat{V}$$
where $\hat{T}$ is the _kinetic energy operator_ and $\hat{V}$ is the _potential energy operator_, which we will construct in turn below.

### Constructing the Kinetic Energy Operator

The electron on our 1-dimensional length of wire will move in a linear trajectory with kinetic energy
$$T = \frac{1}{2}mv^2 = \frac{1}{2}\frac{(mv)^2}{m} = \frac{p^2}{2m}.$$
The quantum mechanical operator corresponding to the classically observable linear momentum in the $x$ direction is
$$\hat{P}_x = -i\hbar\frac{\partial}{\partial x},$$
where $\hbar = \frac{h}{2\pi}$ is the _angular Planck constant_. Because our system is one dimensional, the wavefunction $\Psi$ will depend only on $x$ and the partial derivative in the momentum operator reduces to an ordinary derivative. Substituting $\hat{P}_x$ into the linear kinetic energy gives us the form of the kinetic energy operator in quantum mechanics:
$$\hat{T} = \frac{1}{2m}\hat{P}_x^2 = \frac{1}{2m}\left(-i\hbar\frac{\partial}{\partial x}\right)^2 = \frac{1}{2m}\left(-i\hbar\frac{d}{dx}\right)^2 = \frac{-\hbar^2}{2m}\frac{d^2}{dx^2}$$

### Constructing the Potential Energy Operator

The form of the potential operator $\hat{V}$ in quantum mechanics is defined as "multiply to the right by the potential energy function," which for our one-dimensional model system will be
$$\hat{V} = V(x)\times$$
To define our potential operator $\hat{V}$, we must therefore construct $V(x)$ such that it recovers the following essential characteristics of our real system:

1. The wire is a perfect conductor
    * Because there is no resistance to electron flow, the electron is free to explore the entire length of the wire (i.e., $0<x<L$). To model this behavior, we shall set $V(0<x<L) = 0$.
2. The space beyond the wire is a perfect insulator
    * Because the space surrounding the wire is a perfect insulator, it is impossible for the electron to travel beyond the ends of the wire, no matter how much kinetic energy it may posess. Classically, any finite potential barrier may be overcome by sufficient kinetic energy, so to forbid the particle in our model system from venturing beyond the ends of our "wire," we must therefore set $V(-\infty<x<0) = V(L<x<\infty) = \infty$.

Now that we have defined our model system to recover the behavior of the electron both _on_ the wire and _beyond_ the wire, let us consider what at the _boundary_ of the wire, i.e., at $x=0$ and $x=L$. If the potential energy $V(0)=V(L)=0$, our particle will be allowed to venture arbitrarily close to the forbidden region beyond the wire and its total energy will diverge near these boundaries, which is not physically reasonable. To avoid this we must have that $V(0)=V(L)=\infty$, whereby our potential function is now defined as 

$$V(x) = \begin{cases} 0 & 0 < x < L\\ \infty & {\rm otherwise.}\end{cases}$$

Our model system of a particle subject to this "infinite square well" potential, visualized in [](#fig:1dbox-potential), is often called the "particle in an infinite 1-dimensional box" (or simply the particle-in-a-box, PiB) because it resembles a box with a flat bottom and infinitely tall walls. The potential operator $\hat{V}$ is therefore given by
$$\hat{V} = V(x)\times = \begin{cases}0\times & 0 < x < L\\ \infty\times & {\rm otherwise}\end{cases}$$

### Constructing the Hamiltonian & Defining the PiB TIWE

Now that we have expressions for $\hat{T}$ and $\hat{V}$, we may construct our Hamiltonian $\hat{H}$ and define the time-independent wave equation (TIWE) for our particle-in-a-box model system. The full Hamiltonian is given by
$$\hat{H} = \hat{T} + \hat{V} = \begin{cases} \frac{-\hbar^2}{2m}\frac{d^2}{dx^2} & 0<x<L\\ \infty & {\rm otherwise},\end{cases}$$
however because we have already argued that our particle cannot venture outside the box, we may neglect these forbidden regions and write
$$\hat{H} = \frac{-\hbar^2}{2m}\frac{d^2}{dx^2};\hspace{1cm} 0<x<L$$
This Hamiltonian defines the TIWE
$$\hat{H}\psi = \left[\frac{-\hbar^2}{2m}\frac{d^2}{dx^2}\right]\psi = E\psi.$$

### General Solution to the PiB TIWE

The TIWE above can be rearranged as
$$\frac{d^2\psi}{dx^2} + k^2\psi = 0;\hspace{1cm} k = \frac{\sqrt{2mE}}{\hbar},$$
which is a linear, second-order ordinary differential equation with constant coefficients. Differential equations of this type are known have general solutions of the form
$$\psi(x) = A\sin{kx} + B\cos{kx}$$
where the prefactors $A$ and $B$ are real constants and the frequency factor $k=\frac{\sqrt{2mE}}{\hbar}$ is a positive real number.

### Particular Solutions Arise from the Application of Boundary Conditions

Now that we have demonstrated that the general solution $\psi(x)$ indeed satisfies the TIWE for this system, we must ensure that $\psi(x)$ also satisfies the physical requirements of a well-defined wavefunction collected into the so-called _Born Conditions_:

:::{note} ℹ️ Born Conditions for a Well-Defined Wavefunction
:icon:false
For a wavefunction $\Psi(\tau)$ to be considered well-defined and physically meaningful, the following conditions must be satidfied:
1. $\Psi(\tau)$ must be a single-valued function, i.e., for every collection of variable inputs $\tau$, there exists a single unique output $z$ of the function such that $z = \Psi(\tau)$.
2. $\Psi(\tau)$ must be square-integrable, i.e., the integral of the square modulus of the wavefunction over all space must be strictly finite:
$$\underset{\rm all\ space}{\int}d\tau \left\vert\Psi(\tau)\right\vert^2 = \underset{\rm all\ space}{\int}d\tau\ \Psi^{\ast}(\tau)\Psi(\tau) < \infty$$
3. $\Psi(\tau)$ must be continuous everywhere
4. $\Psi(\tau)$ must have a continuous first derivative, i.e., $\Psi(\tau)$ must be differentiable everywhere[^wfn-differentiability]

[^wfn-differentiability]: Technically, if we allow ourselves to only consider potentials which have a finite number of discontinuities, then the wavefunction $\Psi(\tau)$ is allowed to have a finite number of points in its domain where it is not differentiable.

Clearly, our general solution $\psi(x)$ indeed satisfies the first Born condition because it is already a "proper" mathematical function. Let us momentarily skip over the second Born condition that $\psi(x)$ be square-integrable, and consider whether our general solution $\psi(x)$ is continuous everywhere.

Because the particle is prevented from leaving the box by the infinite potential barrier, we must have that the wavefunction goes to zero in these forbidden regions:
$$\psi(x) = \begin{cases} A\sin{kx} + B\cos{kx} & 0 \leq x \leq L\\ 0 & {\rm otherwise}\end{cases}$$
Because $\psi(x)$ must be continuous over the entire real line $-\infty\leq x\leq\infty$, $\psi(x)$ must also vanish at the barriers of the potential well, i.e., $\psi(0) = \psi(L) = 0$.
